```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("Embalses High Volume") \
    .config("spark.plugins", "org.apache.gravitino.spark.connector.plugin.GravitinoSparkPlugin") \
    .config("spark.sql.gravitino.uri", "http://gravitino:8090") \
    .config("spark.sql.gravitino.metalake", "embalses_metalake") \
    .getOrCreate()
```

```python
spark.sql("USE `embalses_postgres`")
spark.sql("USE DATABASE `public`")
```

```python
spark.sql("""
INSERT INTO embalses_high_volume
SELECT embalse, COUNT(*) AS dias_alto_volumen
FROM `embalses_mysql`.`embalses_db`.`embalses`
WHERE porcentaje_volumen_embalsado > 90
GROUP BY embalse
""")

print("ETL completa: datos insertados usando Gravitino")

spark.stop()
```
